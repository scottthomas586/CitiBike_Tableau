```python
#merge all dataframes
merged_df = pd.concat([df201903, df201904, df201905, df201906, df201907, df202003, df202004, df202005, df202006, df202007])
merged_df
```

```python
#change usertype Subscriber & Customer to Pass User & Annual Member
merged_df['usertype'] = merged_df['usertype'].replace({'Subscriber': 'Annual Member', 'Customer': 'Pass User'})
merged_df
```

```python
#change gender to unknown, male, female
merged_df['gender'] = merged_df['gender'].replace({0: 'Unknown', 1: 'Male', 2: 'Female'})
merged_df
```

```python
#rename the columns
merged_df_columns = merged_df.rename(columns = {'tripduration': 'Trip Duration (in seconds)', 'starttime': 'Start Time', 'stoptime': 'Stop Time', 'start station id': 'Start Station Id',
                                               'start station name': 'Start Station Name', 'start station latitude': 'Start Station Latitude', 'start station longitude': 'Start Station Longitude',
                                               'end station id': 'End Station ID', 'end station name': 'End Station Name', 'end station latitude': 'End Station Latitude',
                                               'end station longitude': 'End Station Longitude', 'bikeid': 'Bike ID', 'usertype': 'Usertype', 'birth year': 'Birth Year', 'gender': 'Gender'})
merged_df_columns
```

```python
#create a trip duration in minutes column
merged_df_columns['Trip Duration (in minutes)'] = round(merged_df_columns['Trip Duration (in seconds)'] / 60, 2)
merged_df_columns
```

```python
#change start time and stop time to datetime type
merged_df_columns['Start Time'] = pd.to_datetime(merged_df_columns['Start Time'])
merged_df_columns['Stop Time'] = pd.to_datetime(merged_df_columns['Stop Time'])
print(merged_df_columns.dtypes)
```

```python
#create a column for the month - year
merged_df_columns['Month / Year'] = merged_df_columns['Start Time'].apply(lambda x: x.strftime('%B-%Y'))
merged_df_columns
```

```python
#create a colunn for the year
merged_df_columns['Year'] = merged_df_columns['Start Time'].apply(lambda x: x.strftime('%Y'))
merged_df_columns
```

```python
print(merged_df_columns.dtypes)
```

```python
#convert year column to an int
merged_df_columns['Year'] = pd.to_numeric(merged_df_columns['Year'], errors='coerce')
merged_df_columns['Year'] = merged_df_columns['Year'].astype(np.int64)
print(merged_df_columns.dtypes)
```

```python
merged_df_columns['Age'] = merged_df_columns['Year'] - merged_df_columns['Birth Year']
merged_df_columns
```

```python
Generation = []
```

```python
for year in merged_df_columns['Birth Year']:
    if year > 1994:
        Generation.append("Gen Z")
    elif year > 1976:
        Generation.append("Millennial")
    elif year > 1964:
        Generation.append("Gen X")
    elif year > 1945:
        Generation.append("Boomer")
    elif year > 1927:
        Generation.append("Silent Generation")
    else:
        Generation.append("Greatest Generation")
```

```python
merged_df_columns['Generation'] = Generation
merged_df_columns
```
